```python
inter = ['BedroomAbvGr_log:secondFlrSF_log', 'BuiltAge:secondFlrSF_log', 'GarageCars:KitchenAbvGr_1', 'GarageCars:firstFlrSF_log', 'GarageCars:secondFlrSF_log',
         'LotArea_log:firstFlrSF_log',  'LotArea_log:secondFlrSF_log', 'firstFlrSF_log:secondFlrSF_log']

variables= '+'.join(list(x_train.columns.difference(['GarageArea_log', 'TotRmsAbvGrd_log', 'LotFrontage_log']))) +'+'+ '+'.join(inter)

lm = ols(f'price ~ {variables}', data=train_data).fit()
lm.summary()
```

```python
print('''
2차 교호작용을 포함한 회귀식을 생성하고, 선형 회귀 분석을 수행한 후, Backward selection으로 변수를 제거하였다.
먼저, 1차항 중에서 GarageArea_log, TotRmsAbvGrd_log가 유의하지 않은 변수로 판단되어, 이를 제거하고 다시 회귀분석을 수행하였다. 1차항의 제거로, 관련된 변수가 포함된 교호작용항도 제거되었다.
교호작용항의 경우 유의하지 않은 변수가 더 많은 관계로, 유의한 변수들만 리스트로 뽑아서 선택하였다. 
선택 이후, LotFrontage_log 변수가 유의하지 않은 변수로 파악되어 추가로 제거하였다. 교호작용항 중, GarageCars:LotArea_log 변수도 유의하지 않아 제거하였다.
변수 선택 이후, 모든 변수가 유의함을 확인하였고, 변수의 개수는 주효과 변수 8개와 교호작용 변수 8개였다.
유의한 교호작용항은 BedroomAbvGr_log:secondFlrSF_log, BuiltAge:secondFlrSF_log, GarageCars:KitchenAbvGr_1, GarageCars:firstFlrSF_log, GarageCars:secondFlrSF_log, LotArea_log:firstFlrSF_log,
LotArea_log:secondFlrSF_log, firstFlrSF_log:secondFlrSF_log이다.
''')
```

### 1-4 벌점, 앙상블을 포함하여 모형에 적합한 기계학습 모델 3가지를 제시하라
(평가지표는 MSE, MAPE, R2 모두 확인할 것)

```python
from sklearn.linear_model import Ridge, Lasso
from sklearn.ensemble import RandomForestRegressor

from sklearn.metrics import mean_squared_error, r2_score
```

```python
ridge = Ridge()
lasso = Lasso()
rf = RandomForestRegressor()
```

```python
def MAPE(y_test, y_pred):
    mape = np.mean(np.abs((y_test - y_pred)/y_test)) * 100
    return mape
```

```python
model_list = [ridge, lasso, rf]

for model in model_list:
    model.fit(x_train_scaled, y_train)
    pred = model.predict(x_test_scaled)

    mse = mean_squared_error(y_test, pred)
    r2 = r2_score(y_test, pred)
    mape = MAPE(y_test, pred)

    print(f"{str(model.__class__()).split('(')[0]} 모델의 MSE는 {mse}, MAPE는 {mape}, R2 score는 {r2}이다.")
```

```python
print('''
테스트 데이터에 대하여 가장 성능이 좋은 모델은 RandomForest 회귀 모델이며, MSE와 MAPE가 다른 모델에 비해 낮으면서 R2 스코어가 0.77로 가장 높게 나타났다.
''')
```

### 2-1 마지막 일자를 기준으로 인구 대비 확진자 비율이 높은 상위 5개 국가를 구하여라
상위 5개 국가별로 누적 확진자, 일일 확진자, 누적 사망자, 일일 사망자, 그래프, 범례를 이용해서 가독성 있게 만들어라     

```python
new_df = pd.DataFrame()

for location in df.location.unique():
    tmp_df = df[df.location==location]

    tmp_df['new_cases'] = tmp_df['total_cases'] - tmp_df['total_cases'].shift(1)
    tmp_df['new_deaths'] = tmp_df['total_deaths'] - tmp_df['total_deaths'].shift(1)
    tmp_df = tmp_df.loc[1:]

    new_df = pd.concat([new_df, tmp_df])
```

```python
new_df
```

```python
df_latest = new_df[new_df.date == new_df.date.max()]
```

```python
df_latest.isna().sum()
```

```python
df_latest = df_latest[df_latest['population'].notna()]
df_latest['case_ratio'] = df_latest['new_cases']/df_latest['population']
df_latest = df_latest.sort_values(['case_ratio'], ascending=False).reset_index(drop=True)
top_5 = df_latest.head(5)
```

```python
location = list(top_5.location.values)
target = new_df[new_df.location.isin(location)].reset_index(drop=True)
target['date'] = pd.to_datetime(target['date'])
```

```python
print('누적 확진자와 누적 사망자의 경우 column에 있으나, 일일 확진자와 일일 사망자의 경우 column으로 제시되어 있지 않다. 누적 변수를 통해서 일일 변수를 생성한다.')
print('각 location 별로 누적 확진자 변수의 shift를 구한 다음, 누적 확진자 수에서 shift를 빼준다.')
print('그리고 Population의 NA 값을 처리한다. 같은 location이면 모두 같은 값을 가지고 있으므로, 해당 값으로 치환한다.')
print('다음으로, 마지막 일자에 대한 데이터만 추출한다. 마지막 일자 데이터에 population 결측치가 존재하는데, 이는 지역이 인구수를 구할 수 없는 international 이므로 제거한다.')
print('2021년 11월 30일 기준, 인구 수 대비 일일 확진자 비율 상위 5개 국가는 Andorra, Czechia, Liechtenstein, San Marino, Slocakia이고, 이들 국가에 대하여 누적 확진, 누적 사망, 일일 확진, 일일 사망 그래프를 그리면 아래와 같다.')
```

설문조사 데이터      
데이터 출처 : 자체 제작    
데이터 경로 : /kaggle/input/adp-kr-p3/problem3.csv 

데이터 설명 : A ~ D까지의 그룹에게 각각 같은 설문조사를 하여 1-1,1-2,1-3...5-1,5-4 인 설문지를 푼 것이다. 문항은 영역별로 나뉘어 있고, 영역은 크게 5개이다(1~5)    
각 영역의 세부문항은 4개씩 존재한다 (1-1,1-2,1-3,1-4 ~) 이 때 중간에 반대 문항이 들어가 있다. 예를 들어 1-1 문제가 "나는 시간약속을 잘 지킨다."라는 문제라면 1-3의 문제는 "나는 시간약속을 잘 지키지 않는다." 라는 역문제로 구성 되어있다. 각 영역의 3번문항은 1번문항의 역문제이다. 
모든 답변은 5점 척도이다. 문제를 풀기전 모든 역문항의 경우 점수를 변환(6점을 빼서) 작업이 필요하다

### 3-1 역문항을 변환 한 후 각 그룹`(A~D)`의 영역`(Q1~Q5)`별 응답의 평균, 표준편차, 왜도, 첨도를 구하라. (각 통계량 별로 4x5 dataframe 생성)

```python
import pandas as pd
df =pd.read_csv('https://raw.githubusercontent.com/Datamanim/datarepo/main/adp/17/problem3.csv')
df.head()
```

```python
Q1_list = ['Q1-1', 'Q1-2', 'Q1-3', 'Q1-4']
Q2_list = ['Q2-1', 'Q2-2', 'Q2-3', 'Q2-4']
Q3_list = ['Q3-1', 'Q3-2', 'Q3-3', 'Q3-4']
Q4_list = ['Q4-1', 'Q4-2', 'Q4-3', 'Q4-4']
Q5_list = ['Q5-1', 'Q5-2', 'Q5-3', 'Q5-4']
```

```python
df.T.loc[['group'] + Q1_list]
```

```python
for col in df.columns:
    if col.endswith('3'):
        df[col] = -df[col]
        df[col] = np.where(df[col]==-5, 1, np.where(df[col]==-4, 2, np.where(df[col]==-3, 3, np.where(df[col]==-2, 4, 5))))
```

```python
import scipy.stats as stats
```

```python
print('역문항을 처리한 다음, 각 그룹별, 영역별 평균, 표준편차, 왜도, 첨도를 계산하면 아래와 같다.')
```

```python
for method in ['mean', 'std', 'skew', 'kurt']:
    m_list = list()
    for i, g in enumerate(['A', 'B', 'C', 'D']):
        g_subset = df[df.group==g]
        m_by_q_list = list()
        for j, q in enumerate([Q1_list, Q2_list, Q3_list, Q4_list, Q5_list]):
            data = np.array(g_subset.loc[:, q]).reshape(1, -1)[0]
            if method=='mean':
                m_by_q_list.append(data.mean())
            elif method=='std':
                m_by_q_list.append(data.std())
            elif method=='skew':
                m_by_q_list.append(stats.skew(data, bias=False))
            elif method=='kurt':
                m_by_q_list.append(stats.kurtosis(data, bias=False))
        m_list.append(m_by_q_list)
    print(f'DataFrame for {method}')
    print(pd.DataFrame(m_list, columns=['Q1', 'Q2', 'Q3', 'Q4', 'Q5'], index=['A', 'B', 'C', 'D']))
    print()
```

### 3-2 그룹별로 Q1-1문항의 차이가 존재하는지 anova분석을 시행하라

```python
aov_data = df.loc[:, ['group', 'Q1-1']]
```

```python
ga = aov_data[aov_data.group=='A']
gb = aov_data[aov_data.group=='B']
gc = aov_data[aov_data.group=='C']
gd = aov_data[aov_data.group=='D']
```

```python
print(stats.shapiro(ga['Q1-1']))
print(stats.shapiro(gb['Q1-1']))
print(stats.shapiro(gc['Q1-1']))
print(stats.shapiro(gd['Q1-1']))
```

```python
print('ANOVA 분석의 가정 중 하나인 각 집단의 정규성을 확인하기 위해, Shapiro-Wilk 검정을 수행하였다. 수행 결과, 모든 집단에서 데이터가 정규성을 따른다는 귀무가설을 기각할 만한 충분한 통계적 근거가 있는 것으로 나왔다.')
print('각 집단의 정규성을 보장할 수 없기 때문에, 비모수적 검정인 Kruskal-Wallis 검정을 수행한다.')
```

```python
stats.kruskal(ga['Q1-1'], gb['Q1-1'], gc['Q1-1'], gd['Q1-1'])
```

```python
print('분석 결과, 검정 통계량 4.57, p-value 0.21로 유의수준 0.05 하에서 귀무가설을 기각할 수 없다. 따라서 각 그룹 별로, Q1-1 문항의 차이가 존재한다고 볼 통계적 근거가 부족하다.')
```
